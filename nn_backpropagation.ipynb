```python
X = 1.2
y = 0.4
learning_rate = 0.1
weight = 0.7
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_loss_activation(target):

  # Range of output activations (y_dash)
  y_dash = np.linspace(0, 1 , 100)  # 100 points from 0 to 1

  # Calculate loss for each y_dash value
  ls = (y_dash - target)**2

  # Calculate derivative of loss (cost function)
  dv = 2 * (y_dash - target)

  # Plot the loss and its derivative vs. output activation
  plt.plot(y_dash, ls, label="Loss")
  plt.plot(y_dash, dv, label="Derivative of Loss")

  # Label axes and title
  plt.xlabel("Output Activation (y_dash)")
  plt.ylabel("Loss/Derivative")
  plt.title("Loss vs. Output Activation (Quadratic Loss) & Derivative")

  # Add legend
  plt.legend()

  # Show the plot
  plt.grid(True)
  plt.show()

plot_loss_activation(y)
```

```python
def plot_loss_weight(X, y):
  # Define a range of weights to evaluate
  weights = np.linspace(0, 1, 100)

  # Calculate activation (a) for each weight
  activation = X * weights

  # Calculate loss for each weight
  losses = [(a - y) ** 2 for a in activation]

  # Calculate derivative of loss (using chain rule)
  derivative = 2 * (X * (weights) - y)  # This is the correction

  # Plot the loss curve and derivative
  plt.plot(weights, losses, label="Loss")
  plt.plot(weights, derivative, label="Derivative of Loss")
  plt.xlabel("Weight")
  plt.ylabel("Loss/Derivative")
  title = "Loss Curve (a - y)^2 and Derivative (notice the intersection at minimum weight)"
  plt.title(title)
  plt.grid(True)

  plt.legend()

  plt.show()

plot_loss_weight(X, y)
```
